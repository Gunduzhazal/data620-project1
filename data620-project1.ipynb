```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
```

```python
airport = pd.read_csv("Project1_data.csv")
airport.head(10)
```

```python
plt.figure(figsize=(16, 16))

# Create Directional graph
flight_graph = nx.from_pandas_edgelist(airport, source="Nodes_1", target="Nodes_2", create_using=nx.Graph())

print(nx.info(flight_graph))
```

Degree centrality is a measure of how many neighbors a node has. 
With this advantage, nodes with more edges can have multiple ways to access resources to achieve goals.

Eigenvector centrality measures a node’s importance based on the importance of its neighbors. 
Unlike degree centrality, it does not take into account how many connections a node has. 

That is, a person with several connections may have a higher measure of eigenvector centrality depending on 
how well these connections are connected.

```python
tbl_centrality = pd.DataFrame()
tbl_centrality["degree_centrality"] = nx.degree_centrality(flight_graph).values()
tbl_centrality["eigenvector_centrality"] = nx.eigenvector_centrality(flight_graph).values()
tbl_centrality.index = list(flight_graph)
tbl_centrality
```
